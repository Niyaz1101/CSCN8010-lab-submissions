```python
#Libraries

import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".2")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

Bivariate plot with multiple elements(scatterplot, data point and density contours) drawn using the seaborn and matplotlib libraries in Python.

```python
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries'
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

A pie chart using the seaborn and matplotlib libraries in Python. The data is from the Gapminder dataset, specifically for the year 2007 and the European continent.

```python
sns.set_theme(style="ticks")

penguins = sns.load_dataset("penguins")
g = sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde",
)
```

The plot uses density estimation to smoothly represent the distribution of data points. The plot represents bill length and bill depth for penguins, with each species represented by a different color

```python
sns.set_theme(style="darkgrid")

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

Seaborn library is used to plot a line graph of the responses for different events and regions in an example dataset of functional magnetic resonance imaging (fmri) data.
